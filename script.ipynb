```python
import pandas as pd  
df = pd.read_excel("Files/Maestro de trabajadores cierre septiembre.xlsx", skiprows=3)  
dirty_data = df[['Nomre y apellidos', 'CI', 'Unidad']]
#df.head(50)

unidades = []
for index, fila in dirty_data.iterrows():
    persona = {
        'nombre': fila['Nomre y apellidos'],
        'ci': fila['CI'], 
        'unidad': fila['Unidad']
    }
    unidades.append(persona)

print(f"Total de registros: {len(unidades)}")
print(unidades[:5])  # Mostrar solo los prime

```
